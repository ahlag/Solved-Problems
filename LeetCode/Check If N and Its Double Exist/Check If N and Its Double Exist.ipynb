```python
def checkIfExist(arr) -> bool:

    existence = set()
    n = len(arr)

    for i in range(n):
        
        if arr[i]/2 in existence or  arr[i] * 2 in existence:
            return True
        else:
            existence.add(arr[i])

    return False
```

```python
arr = [3,1,7,11]

checkIfExist(arr)
```

```python
def checkIfExist(arr) -> bool:
    hashset = {v:k for k,v in enumerate(arr)}
    for a in range(len(arr)):
        if 2*arr[a] in hashset and hashset[2*arr[a]]!=a:
            return True
    return False
```

```python
arr = [3,1,7,11]

checkIfExist(arr)
```
