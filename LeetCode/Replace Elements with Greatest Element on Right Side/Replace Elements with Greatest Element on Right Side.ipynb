```python
## O(n^2) Time
def replaceElements(arr):
        
    n = len(arr)
    for i in range(n-1):
        arr[i] = max(arr[i+1:n])
        
    arr[-1] = -1
        
    return arr
```

```python
arr = [17,18,5,4,6,1]

replaceElements(arr)
```

```python
## O(n) Time
def replaceElements(arr):
        
    n = len(arr)
    largest = arr[-1]
    arr[-1] = -1
    
    for i in reversed(range(n-1)):
        tmp = arr[i]
        arr[i] = largest
        
        if largest < tmp:
            largest = tmp
        
    return arr

arr = [17,18,5,4,6,1]

replaceElements(arr)
```
