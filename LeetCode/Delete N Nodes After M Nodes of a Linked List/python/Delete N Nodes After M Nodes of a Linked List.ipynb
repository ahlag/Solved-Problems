```python
def deleteNodes(head, m: int, n: int):
    
    curr = head
    
    while curr:
        count = 0
        while count < m and curr:
            prev = curr
            curr = curr.next
            count += 1
        
        temp = curr
        count = 0
        while count < n and temp:
            temp = temp.next
            count += 1
        
        prev.next = temp
        curr = temp
            
    return head
```
