```python
class Solution:
    def generateTrees(self, n: int) -> List[Optional[TreeNode]]:
        
        @lru_cache()
        def dfs(st, end):
            if st > end:
                return [None]
            ans = []
            for i in range(st, end+1):
                left = dfs(st, i-1)
                right = dfs(i+1, end)
                for l in left:
                    for r in right:
                        root = TreeNode(i)
                        root.left = l
                        root.right = r
                        ans.append(root)
            return ans
        
        return dfs(1,n)
```
