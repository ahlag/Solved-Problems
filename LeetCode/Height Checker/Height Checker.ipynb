```python
import copy

def heightChecker(heights) -> int:
        
        
    n = len(heights)
    cnt = 0
        
    expected = sorted(copy.deepcopy(heights))
        
    for i in range(n):
            
        if heights[i] != expected[i]:
            cnt += 1
                
    return cnt
```

```python
heights = [1,2,3,4,5]

heightChecker(heights)
```
