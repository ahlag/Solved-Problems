```python
def using_bit_manipulation(n):
    count = 0
    while n:
        if n & 1: count += 1
        n = n >> 1
    return count
```

```python
def hammingWeight(self, n):
    
    counter = 0

    while n:
        
        # this will take out the right-most 1 of n    
        n = n & (n-1)
        
        # update counter
        counter += 1
    
    return counter
```
