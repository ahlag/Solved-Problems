```python
def sortArrayByParity(nums):
    
    n = len(nums)
    
    for i in range(n-1):
        
        if nums[i] % 2 != 0:
            
            j = n - 1
            while j > i:
                
                if nums[j] % 2 == 0:
                    break
                    
                j -= 1
                    
            nums[i], nums[j] = nums[j], nums[i]
            
    return nums
```

```python
nums = [3,1,2,4]

sortArrayByParity(nums)
```

```python
def sortArrayByParity(nums):
    
    i, j = 0, len(nums) - 1
    
    while i < j:
        
        if nums[i] % 2 > nums[j] % 2:
            nums[i], nums[j] = nums[j], nums[i]
            
        if nums[i] % 2 == 0:
            i += 1
        if nums[j] % 2 != 0:
            j -= 1
            
    return nums

nums = [3,1,2,4]

sortArrayByParity(nums)
```
