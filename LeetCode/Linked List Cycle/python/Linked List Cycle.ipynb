```python
def hasCycle(head):
    nodes_seen = set()
    while head is not None:
        if head in nodes_seen:
            return True
        nodes_seen.add(head)
        head = head.next
    return False
```

```python
def hasCycle(head):
    if head is None:
        return False
    slow = head
    fast = head.next
    while slow != fast:
        if fast is None or fast.next is None:
            return False
        slow = slow.next
        fast = fast.next.next
    return True
```
