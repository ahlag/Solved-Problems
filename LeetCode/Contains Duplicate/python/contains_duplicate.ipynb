```python
def containsDuplicate(nums) -> bool:
    
    nums = sorted(nums)
    
    for i in range(len(nums) - 1):
        if nums[i] == nums[i+1]:
            return True
    
    return False

containsDuplicate([1,2,3,1])
```

```python
def containsDuplicate(nums) -> bool:
    
    seen = set()
    
    for num in nums:
        if num not in seen:
            seen.add(num)
        else:
            return True
    
    return False

containsDuplicate([1,2,3,1])
```
