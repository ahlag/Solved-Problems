```python
def thirdMax(nums): 
    
    nums = set(nums)
    
    maximum = max(nums)
    
    if len(nums) < 3:
        return maximum
    
    nums.remove(maximum)
    second_maximum = max(nums)
    nums.remove(second_maximum)
    
    return max(nums)

nums = [3,2,1]
thirdMax(nums)
```

```python
def thirdMax(nums): 
    
    def maximum_ignoring_seen_maximums(nums, seen_maximums):
        maximum = None
        
        for num in nums:
            if num in seen_maximums:
                continue
            if maximum == None or num > maximum:
                maximum = num
        return maximum
    
    seen_maximums = set()
    
    for _ in range(3):
        current_maximum = maximum_ignoring_seen_maximums(nums, seen_maximums)
        if current_maximum == None:
            return max(seen_maximums)
        seen_maximums.add(current_maximum)

    return min(seen_maximums)

nums = [3,2,1]
thirdMax(nums)
```

```python
def thirdMax(nums): 
    
    maximas = set()
    
    for num in nums:
        maximas.add(num)
        if len(maximas) > 3:
            maximas.remove(min(maximas))
                
    if len(maximas) == 3:
        return min(maximas)

    return max(maximas)

nums = [3,2,1]
thirdMax(nums)
```
