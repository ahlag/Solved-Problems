```python
def findDisappearedNumbers(nums):
    
    n = len(nums)
    seen = set()
    
    for num in nums:
        if num not in seen:
            seen.add(num)
            
    return [x for x in range(1,n+1) if x not in seen]

nums = [4,3,2,7,8,2,3,1]

findDisappearedNumbers(nums)
```

```python
def findDisappearedNumbers(nums):
    
    hash_table = {}
    
    for num in nums:
        hash_table[num] = 1
        
    result = []
    
    for num in range(1, len(nums) + 1):
        if num not in hash_table:
            result.append(num)
        
    return result

nums = [4,3,2,7,8,2,3,1]

findDisappearedNumbers(nums)
```

```python
def findDisappearedNumbers(nums):
    
    n = len(nums)
    
    for i in range(n):
        
        new_index = abs(nums[i]) - 1
        
        if nums[new_index] > 0:
            nums[new_index] *= -1
    
    result = []
    
    for i in range(n):
        if nums[i] > 0:
            result.append(i+1)
            
    return result

nums = [4,3,2,7,8,2,3,1]

findDisappearedNumbers(nums)
```
