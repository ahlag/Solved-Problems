```python
def validMountainArray(arr) -> bool:
    
        up = down = False
        n = len(arr)
        
        max_index = arr.index(max(arr))
        
        for i in range(1, max_index+1):

            if arr[i] - arr[i-1] > 0:
                up = True
            else:
                up = False
                break
                
        for i in range(max_index+1, n):
                
            if arr[i] - arr[i-1] < 0:
                down = True
            else:
                down = False
                break
                
        return up and down
```

```python
def validMountainArray(arr) -> bool:
        
    n = len(arr)
    i = 0
    
    # walk up
    while i+1 < n and arr[i] < arr[i+1]:
        i += 1
        
    # peak can't be first or last
    if i == 0 or i == n-1:
        return False
    
    # walk down
    while i+1 < n and arr[i] > arr[i+1]:
        i += 1
        
    return i == n-1
```

```python
arr = [0,3,2,1]

validMountainArray(arr)
```
